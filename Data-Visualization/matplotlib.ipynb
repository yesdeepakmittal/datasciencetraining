```python
import matplotlib.pyplot as plt
```

### Pie chart

```python
sizes = [3,5,2]
labels = 'natural language processing'.split()
```

```python
plt.pie(sizes, labels=labels, autopct='%1.2f%%',shadow=True, startangle=90)
plt.show()
```
