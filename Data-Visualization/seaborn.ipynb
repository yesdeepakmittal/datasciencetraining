- Built on the top of matplotlib
- Works well with pandas dataframe

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

# Univariated Analysis
- One variable


#### Histogram 

```python
sns.histplot(data=df,x='petal_length',bins=20,hue='species')
ax = plt.twinx()
sns.kdeplot(data = df,x='petal_length',ax=ax,hue='species')
plt.plot()
```

**Observation**
- No flower having petal length b/t 2.2 - 2.9
- bimodal distribution

```python
setosa = df[df.species == 'setosa']
versicolor = df[df.species == 'versicolor']
virginica = df[df.species == 'virginica']

sns.kdeplot(data = setosa,x='petal_length',label='setosa')
sns.kdeplot(data = versicolor,x='petal_length')
sns.kdeplot(data = virginica,x='petal_length')
```

```python
sns.kdeplot(data=df,x='petal_length',hue='species')
```

```python
sns.displot(data=df,x='petal_length',hue='species')
```

# Bivariate Analysis

```python
sns.scatterplot(data=df,x='petal_length',y='petal_width',hue='species')
```

```python
# Joint Plot
sns.jointplot(data=df,x='sepal_length',y='sepal_width',kind='reg')
```

```python
sns.jointplot(data=df,x='petal_width',y='petal_length',kind='scatter',hue='species')
```

# Hexplot
Good if there is a high conc. of data points in a region

```python
sns.jointplot(data=df,x='sepal_length',y='sepal_width',kind='hex')
```

# Multivariate Analysis

### Pairplot

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(data=df,hue='species')
```

- Which two features can separate the category-wise data points very clearly? 
-- petal width
-- petal length

```python
# countplot
sns.countplot(data=df,x='species')
```
