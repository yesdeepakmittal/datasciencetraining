In this case study we will find out the number of specfic size of T-shirt a marketing should order such that the target receivers have different heights

Assumptions
- Height is a continuous random variable
- Normal Distribution will be used

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import math
%matplotlib inline
```

```python
# Lets create data for 10000 people heights with mean 164.67 & std. 6.888
N = 10000
MEAN = 164.67
STD = 6.888

distribution = stats.norm(loc=MEAN,scale=STD)
data = distribution.rvs(N)
sns.histplot(data, kde =True, stat='density')
plt.show()
```

```python
# CDF 
sns.histplot(data, kde=True, stat='density', cumulative=True, fill=False)
plt.show()
```

```python
# PDF - derivate of CDF
x = np.arange(130,190)

pdf = distribution.pdf(x)
plt.plot(x, pdf)
plt.ylabel("Probability")
plt.grid()
plt.show()
```

```python
# what is the cdf value for x <= 150
'''
z = (150 - 164.67)/6.888 = -2.13
cdf @ z = -2.13 = 0.0166
'''

distribution.cdf(x = 150)
```

```python
# If we have 1000 employees, how many T-shirts do we need to make whose size is upto 150?
np.ceil(distribution.cdf(x = 150) * 1000)
```

```python
# t-shirt b/t 150 - 160
np.ceil((distribution.cdf(x = 160) - distribution.cdf(x = 150))*1000)
```
