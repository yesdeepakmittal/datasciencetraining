The formula for the coefficient of skewness (denoted as "Skewness" or "S") is given by:

$$S = \frac{\frac{1}{n}\sum_{i=1}^n (x_i - \bar{x})^3}{s^3}$$

Where:

$x_i$ is the $i$th data point

$\bar{x}$ is the sample mean

$n$ is the number of data points

$s$ is the sample standard deviation

```python
import numpy as np

arr = np.array([38,36,31,30,20,28,25,26,19,21,19,21,20,21,21,20,21,24,32,35,38,38])
```

```python
def coeff_skewness(arr):
    mean = arr.mean()
    std = arr.std()
    
    x = np.sum((arr - mean)**3)/arr.__len__()

    return x/std**3

coeff_skewness(arr)
```

$CS$ : Positive -> Positively Skewed(Right Skewed)

$CS$ : Negative -> Negatively Skewed(Left Skewed)

$CS$ ~ Zero -> Lesser Degree of Skewness

Above provided data is Right Skewed

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate skewness curve
fit = stats.norm.pdf(arr, np.mean(arr), np.std(arr)) 
plt.plot(arr, fit,'o')

# Add labels and title
plt.xlabel('Data')
plt.ylabel('Probability Density')
plt.title('Skewness Curve of Data')

plt.axvline(x=arr.mean(), color='r', linestyle='--')
plt.text(arr.mean(),0.03,'Mean',rotation=90,va='center')

plt.axvline(x=np.median(arr), color='g', linestyle='--')
plt.text(np.median(arr),0.03,'Median',rotation=90,va='center')

text = 'Positively Skewed'
plt.text(33,0.05,text,rotation=0,va='center',color='indianred')

# Display plot
plt.show()
```
